## test control vars by group

```python
pvals_save =pd.DataFrame()
for group in ['insurance','usage',  'demographics', 'access', 'health_status', 'health_condition_summary','transportation',
       'hearing_vision', 'health_knowledge', 'risk',  'housing']:
    indvars1 = control_vars_all[(control_vars_all['control_var_group'] == group )
        & (control_vars_all['tier2'].isin(['none','male','fem']))]['var_name']
    
    if group == 'demographics':
        options = [['dem_re_age'], ['dem_race' , 'dem_age']]
    elif group == 'insurance':
        options = [['adm_op_mdcd']]
    else:
        options = [ ['No excluded vars']]

    for opt in options:
        indvars2 =   [d for d in indvars1 if d not in opt]
        indvars3 =   [d for d in indvars2 if d not in cat_vars] 
        indcatvars =  [d for d in indvars2 if d  in cat_vars] 

        indcatvar_dummies = []
        for catvar in  indcatvars:
            for v in cat_var_dummies:
                if (v.startswith(catvar)) and not (v.endswith('x')): 
                    indcatvar_dummies.append(v)

        X = sm.add_constant(df[indvars3 + indcatvar_dummies] )
        y = df['y']
        logit_mod = sm.Logit( y, X )
        logit_res = logit_mod.fit()

        print()
        print(group , opt)
        print(logit_res.summary())
        
        pvals = pd.DataFrame(logit_res.pvalues).reset_index()
        pvals.columns = ['var_name','pval']
        pvals['group']=group
        pvals['option'] = opt[0]
        
        pvals_save = pd.concat([pvals_save, pvals])
```
